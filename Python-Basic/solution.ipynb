# **Assignment 1: Temperature Check**

```python
current_temperature=int(input("Enter the current temperature : "))

if current_temperature<10:
  print("It's too cold!")

elif current_temperature>=10 and current_temperature<=25:
  print("It's cool outside")

else:
  print("It's hot outside!")
```

# **Assignment 2: ATM Withdrawal System**

```python
account_balance=int(input("Please Enter Your Account Balance : "))
withdraw_amount=int(input("Please Enter the Amount That You Want to Withdraw : "))

if withdraw_amount>account_balance:
  print("Insufficient Balance")

else:
  print("Transaction Successful")
```

# **Assignment 3: Student Grade Calculator**

```python
marks=int(input("Please Enter your marks between 0-100 : "))

if marks>=90 and marks<=100:
  print("A+")

elif marks>=80 and marks<=89:
  print("A")

elif marks>=70 and marks<=79:
  print("B")

elif marks>=60 and marks<=69:
  print("C")

elif marks>=0 and marks<60:
  print("Fail")

else:
  print("Invalid Marks! Please Enter your marks between 0-100")

```

# **Assignment 4: Bus Ticket Discount**

```python
age=int(input("Please Enter the age : "))

if age<5:
  print("Free")

elif age>=5 and age<=18:
  print("50% Discount")

elif age>=60:
  print("30% Discount")

else:
  print("Full Price")
```

# **Assignment 5: Restaurant Bill with Discount**

```python
total_bill=int(input("what is the total bill : "))

if total_bill > 1000:
  final_bill=total_bill - total_bill*(10/100)
else:
  final_bill=total_bill

print(final_bill)
```

# **Assignment 7: Traffic Light System**

```python
traffic_color=input("Enter the traffic color : red, yellow, or green -> ")

if traffic_color=="red":
  print("Stop")

elif traffic_color=="yellow":
  print("Ready to go")

elif traffic_color=="green":
  print("Go")

else:
  print("Invalid color")
```

# **Assignment 8: Leap Year Checker**

```python
year=int(input("Enter the year : "))

if year%400==0:
  print("Leap Year")

elif year%4==0 and year%100!=0:
  print("Leap Year")

else:
  print("Not a Leap Year")
```

# **Assignment 10: Electricity Bill Calculator**

```python
units=int(input("Please Enter the Number of Units Consumed: "))

if units<=100:
  bill=units*5

elif units>=101 and units<=200:
  bill=units*7

else:
  bill=units*10

print(f'Total Payable Amoutn : {bill}')
```
